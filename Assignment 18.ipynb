```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(r"C:\Users\Administrator\Documents\Data Science Bootcamp\K means clustering\Mall_Customers.csv")
df.head()
```

```python
X=df.iloc[:,[3,4]]
print(X.values)
```

```python
print("\nMissing values:\n", X.isnull().sum())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
print(y_kmeans)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[y_kmeans == 0, 0], X_scaled[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X_scaled[y_kmeans == 1, 0], X_scaled[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X_scaled[y_kmeans == 2, 0], X_scaled[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X_scaled[y_kmeans == 3, 0], X_scaled[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X_scaled[y_kmeans == 4, 0], X_scaled[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Customer Segmentation')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```

```python
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++')
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS Score")
plt.show()
```

```python
optimal_k=5
plt.figure(figsize=(8,6))

# Loop through each cluster
for i in range(optimal_k):
    plt.scatter(
        X_scaled[y_kmeans == i, 0], X_scaled[y_kmeans == i, 1],
        s=100, label=f'Cluster {i+1}'
    )

# Plot the centroids
centroids = kmeans.cluster_centers_
plt.scatter(
    centroids[:, 0], centroids[:, 1],
    s=300, c='black', marker='X', label='Centroids'
)

plt.title('K-Means Clustering of Customers')
plt.xlabel('Annual Income (scaled)')
plt.ylabel('Spending Score (scaled)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for k in [3, 4, 5, 6]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    print(f'\nCentroids for k={k}:\n', pd.DataFrame(kmeans.cluster_centers_, columns=['Annual Income', 'Spending Score']))
```
