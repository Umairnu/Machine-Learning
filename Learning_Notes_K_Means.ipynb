# **K-Means Clustering**

**K-Means** is an unsupervised machine learning algorithm used for clustering. It divides a dataset into **K** distinct, non-overlapping clusters based on feature similarity. The algorithm aims to minimize the variance within each cluster by iteratively adjusting the cluster centroids.

### **Key Concepts**

1. **Cluster Centroids**:
   - The centroid is the center of a cluster, representing the mean of all points within that cluster. It’s recalculated at each iteration.

2. **K-Means Algorithm**:
   - **Step 1**: Initialize K centroids randomly (or using a more advanced method like K-Means++ to improve convergence).
   - **Step 2**: Assign each data point to the nearest centroid.
   - **Step 3**: Recompute the centroids as the mean of all points assigned to each cluster.
   - **Step 4**: Repeat Steps 2 and 3 until convergence (when centroids do not change or change very little).

3. **Elbow Method**:
   - To determine the optimal number of clusters (K), you can use the **Elbow Method**, which looks at the within-cluster sum of squares (WSS) as a function of K. The point where the WSS starts to decrease at a slower rate (the "elbow") is often a good choice for K.

4. **Inertia (within-cluster sum of squares)**:
   - Measures how tight the clusters are. Lower inertia means the clusters are more compact.

### **Advantages**:
- Simple and easy to implement.
- Efficient for large datasets.
- Can work with large numbers of features (high-dimensional data).

### **Disadvantages**:
- Requires the number of clusters (K) to be specified beforehand.
- Sensitive to the initial placement of centroids.
- Can struggle with non-spherical or overlapping clusters.
- Does not perform well with clusters of different sizes or densities.

### **Applications**:
- Customer segmentation in marketing.
- Document clustering in natural language processing.
- Image compression.
- Anomaly detection in security.



## **1. K-Means Clustering on a Simple Dataset**

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Create a simple dataset
X = np.array([[1, 2], [1, 4], [1, 0], [4, 2], [4, 4], [4, 0]])

# Initialize KMeans with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)

# Fit the model
kmeans.fit(X)

# Get the cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Print the centroids and labels
print("Centroids:\n", centroids)
print("Labels:", labels)

# Plot the data points and centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red')
plt.title('K-Means Clustering')
plt.show()
```

**Explanation**:

* **KMeans**: Performs clustering with 2 clusters.
* **Labels**: The cluster assignment for each data point.
* **Centroids**: The center of each cluster.
* **Plot**: Visualizes the data points and their respective cluster centroids.

## **2. Using K-Means on the Iris Dataset**

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
data = load_iris()
X = data.data

# Apply K-Means with 3 clusters (since the Iris dataset has 3 species)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the first two features with cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='red')
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

**Explanation**:

* **Iris dataset**: A well-known dataset with 3 classes of flowers.
* **Clustering**: Divides the dataset into 3 clusters.
* **Visualization**: Plots the first two features of the dataset with the cluster assignments.


## **3. Elbow Method to Determine Optimal K**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
X = data.data

# Calculate the sum of squared distances (inertia) for different values of K
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

**Explanation**:

* **Inertia**: The within-cluster sum of squares, which measures the tightness of clusters.
* **Elbow Plot**: Helps determine the optimal number of clusters by looking for the "elbow" point where inertia starts to decrease at a slower rate.

### **Summary**:

- **K-Means** is a simple and widely used clustering algorithm that groups data points into K clusters.
- The **Elbow Method** helps determine the best value for K.
- K-Means is fast and works well for spherical clusters, but may struggle with clusters of varying shapes, sizes, or densities.
